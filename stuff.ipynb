```python
#Basics + Line Charts

from matplotlib import pyplot as plt 

print(plt.style.available) # To find styles
plt.style.use('ggplot')
#plt.xkcd()  # Comic looking style

x = [1,1,2,3,5,8,13,21]
y = [1,3,5,7,9,11,13,15]
z = [3,8,20,17,23,6,18,17]

plt.plot(x, y, linestyle = "--", color = 'k', marker = '.', linewidth = 1 , label="X and Y")
plt.plot(x, z, linestyle = '-.', color = 'b', marker = 'o', label="X and Z")

plt.title("Test Graph")

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(True)

plt.legend()

plt.tight_layout() # for better padding on some machines

#plt.savefig('plot.png') # Saves in current directory; full path can also be given

plt.show()
```

```python
# line parameters: x, y, colour, linestyle, linewidth, markers, labels
# colours can be hex values as well
# refer format strings
# lines overlap in the order in which they are added
```

```python
# Bar Charts

import numpy as np # Needed for arange function
from matplotlib import pyplot as plt 

plt.style.use('fivethirtyeight')

x = [1,1,2,3,5,8,13,21]
x_indexes = np.arange(len(x))

lwidth = 0.25

y = [1,3,5,7,9,11,13,15]
z = [3,8,20,17,23,6,18,17]

plt.bar(x_indexes - lwidth, y, linestyle = "--", width= lwidth, color = 'k',  label="X and Y") 
plt.bar(x_indexes, z, linestyle = '-.', width= lwidth, color = 'b',  label="X and Z")

plt.title("Test Graph")

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(True)

plt.legend()

plt.xticks(ticks = x_indexes, labels = x) # To get actual values labelled on x axis and not indexes that we used to shift bars around

plt.tight_layout() 

plt.show()
```

```python
# width paramter exists for bar charts
# marker paramter does not exist for bar charts
# numpy is used for arange function which is used to shift bars to left and right
# without using arange and indexes, bar appear on top of each other and information is lost
```

```python
# Reading from CSV Files

import csv
import numpy as np
from collections import Counter
from matplotlib import pyplot as plt 

plt.style.use('fivethirtyeight')

with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file) # returns dictionary 

    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';')) # Takes list of values for "LanguagesWorkedWith" as list

languages = []
popularity = []

#print(language_counter.most_common(15)) # We have data on no. of ppl who know certain languages, sorted w most responses at beginning

for item in language_counter.most_common(15): #loops over the tuples
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title('Most Popular Languages')
plt.ylabel('Number of Users')

#plt.tight_layout()

plt.show()  

```

```python
# csv.DictReader() returns data as an ordered dictionary (key-value pairs)
# Counter() from collections module helps count no. of occurences of every element in a list 
# .barh() makes horizontal bar chart
# expects y axis args first
```

```python
# Reading CSV using pandas

import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt 

plt.style.use('fivethirtyeight')

data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';')) # Takes list of values for "LanguagesWorkedWith" as list

languages = []
popularity = []

#print(language_counter.most_common(15)) # We have data on no. of ppl who know certain languages, sorted w most responses at beginning

for item in language_counter.most_common(15): #loops over the tuples
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title('Most Popular Languages')
plt.ylabel('Number of Users')

#plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter

plt.style.use('seaborn')

covidata = pd.read_csv('COVID_Dataset.csv')

ages = list(covidata.Age)
agecount = Counter(ages)
#print(agecount)
ageord = []
for el in agecount:
     ageord.append(el)
ageord.sort()
finage = []
for a in ageord:
    k = [a,agecount[a]]
    finage.append(k)
#print(finage)
#print(covidata.Age.min())

ages = []
cases = []
for a in finage:
    ages.append(a[0])
    cases.append(a[1])
#print(ages)
#print(cases)

plt.plot(ages, cases)







```
